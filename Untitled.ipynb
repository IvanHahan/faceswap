```python
from scipy.io import loadmat
import dlib
import cv2
import os
import numpy as np
import boto3
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('../shape_predictor_68_face_landmarks.dat')
```

```python
def make_dir_if_needed(dir_):
    if not os.path.exists(dir_):
        os.makedirs(dir_)
```

```python
data_dir = '../aligned_images_DB/'
output_dir = '../faces/'
```

```python
make_dir_if_needed(output_dir)
```

```python
for person in os.listdir(data_dir):
    print(person)
    for video in os.listdir(os.path.join(data_dir, person)):
        for frame_name in os.listdir(os.path.join(data_dir, person, video)):
            frame_path = os.path.join(data_dir, person, video, frame_name)
            frame = cv2.imread(frame_path)
            if frame is None:
                continue
            shapes = detector(frame)
            for s in shapes:

                landmarks = predictor(frame, s)

                landmarks_np = np.zeros((0, 2), dtype='int32')
                for i in range(68):
                    p = landmarks.part(i)
                    landmarks_np = np.append(landmarks_np, [[p.x, p.y]], axis=0)

                top = landmarks_np[:, 1].min() - 10
                bottom = landmarks_np[:, 1].max() + 10
                left = landmarks_np[:, 0].min() - 10
                right = landmarks_np[:, 0].max() + 10
                face_segment = frame[top:bottom, left:right].copy()

                landmarks_np[:, 0] -= left
                landmarks_np[:, 1] -= top

                cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 140), 2)
                for i in range(68):
                    p = landmarks.part(i)
                    cv2.circle(frame, (p.x, p.y), 1, (0, 0, 255), -1)
                    
                output_frame_path = os.path.join(output_dir, person, 'frames')
                output_landmarks_path = os.path.join(output_dir, person, 'landmarks')
                
                make_dir_if_needed(output_frame_path)
                make_dir_if_needed(output_landmarks_path)
                
                landmarks_name = frame_name + '.npy'
                
                cv2.imwrite(os.path.join(output_frame_path, frame_name), face_segment)
                np.save(os.path.join(output_landmarks_path, landmarks_name), landmarks_np)
                
```
